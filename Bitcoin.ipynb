```python
# ---------- DAILY AGGREGATE ACROSS ALL TRADERS ----------
daily = merged.groupby('Date').agg(
    total_pnl=('Closed PnL', 'sum'),
    trade_count=('Closed PnL', 'count'),
    win_count=('Closed PnL', lambda x: (x > 0).sum())
).reset_index()

daily['win_rate_%'] = daily['win_count'] / daily['trade_count'] * 100
daily = daily.merge(df2[['Date', 'classification']], on='Date', how='left').sort_values('Date')

# ---------- DRAWDOWN PROXY ----------
# cumulative PnL over time, distance below running peak
daily['cum_pnl'] = daily['total_pnl'].cumsum()
daily['running_max'] = daily['cum_pnl'].cummax()
daily['drawdown'] = daily['cum_pnl'] - daily['running_max']   # 0 = at peak, negative = below peak

# ---------- PERFORMANCE BY SENTIMENT (5 categories) ----------
order = ['Extreme Fear', 'Fear', 'Neutral', 'Greed', 'Extreme Greed']

by_sentiment = daily.groupby('classification').agg(
    avg_daily_pnl=('total_pnl', 'mean'),
    total_pnl=('total_pnl', 'sum'),
    avg_win_rate=('win_rate_%', 'mean'),
    avg_drawdown=('drawdown', 'mean'),
    worst_drawdown=('drawdown', 'min'),
    num_days=('Date', 'count')
).reindex(order)

print("=== Performance by sentiment ===")
by_sentiment   # last line — displays as table
```

```python
# ---------- PER-TRADER METRICS ----------
daily_trader = merged.groupby(['Account','Date']).agg(daily_pnl=('Closed PnL','sum')).reset_index()

trader_stats = merged.groupby('Account').agg(
    total_trades=('Closed PnL','count'),
    total_pnl=('Closed PnL','sum'),
    win_count=('Closed PnL', lambda x: (x>0).sum()),
    avg_size_usd=('Size USD','mean')
).reset_index()
trader_stats['win_rate_%'] = trader_stats['win_count'] / trader_stats['total_trades'] * 100

daily_std = daily_trader.groupby('Account')['daily_pnl'].agg(
    std_daily_pnl='std', pct_profitable_days=lambda x: (x>0).mean()*100
).reset_index()
trader_stats = trader_stats.merge(daily_std, on='Account', how='left')

# Segment 1: size
size_median = trader_stats['avg_size_usd'].median()
trader_stats['size_segment'] = trader_stats['avg_size_usd'].apply(lambda x: 'High Size' if x >= size_median else 'Low Size')

# Segment 2: frequency
trades_median = trader_stats['total_trades'].median()
trader_stats['freq_segment'] = trader_stats['total_trades'].apply(lambda x: 'Frequent' if x >= trades_median else 'Infrequent')

# Segment 3: consistency
trader_stats['cv'] = trader_stats['std_daily_pnl'] / trader_stats['total_pnl'].abs()
cv_median = trader_stats['cv'].median()
trader_stats['consistency_segment'] = trader_stats['cv'].apply(lambda x: 'Consistent' if x <= cv_median else 'Inconsistent')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.barplot(data=trader_stats, x='size_segment', y='win_rate_%', order=['Low Size','High Size'],
            hue='size_segment', palette=['#4caf50','#e53935'], legend=False, ax=axes[0])
axes[0].set_title('Win Rate: High vs Low Position Size Traders')

sns.barplot(data=trader_stats, x='freq_segment', y='total_pnl', order=['Infrequent','Frequent'],
            hue='freq_segment', palette=['#ff9800','#2e7d32'], legend=False, ax=axes[1])
axes[1].set_title('Avg Total PnL: Frequent vs Infrequent Traders')

sns.barplot(data=trader_stats, x='consistency_segment', y='pct_profitable_days', order=['Inconsistent','Consistent'],
            hue='consistency_segment', palette=['#9e9e9e','#1976d2'], legend=False, ax=axes[2])
axes[2].set_title('% Profitable Days: Consistent vs Inconsistent Traders')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# assumes df1, df2, merged already exist from earlier steps
daily = merged.groupby(['Account', 'Date']).agg(
    daily_pnl=('Closed PnL', 'sum'),
    trade_count=('Closed PnL', 'count'),
    avg_size_usd=('Size USD', 'mean'),
    win_count=('Closed PnL', lambda x: (x > 0).sum()),
    sentiment_value=('value', 'first'),
).reset_index()
daily['win_rate'] = daily['win_count'] / daily['trade_count']

pnl_std = merged.groupby(['Account','Date'])['Closed PnL'].std().reset_index(name='pnl_std_today')
daily = daily.merge(pnl_std, on=['Account','Date'], how='left')
daily['pnl_std_today'] = daily['pnl_std_today'].fillna(0)

daily = daily.sort_values(['Account','Date'])
daily['next_day_pnl'] = daily.groupby('Account')['daily_pnl'].shift(-1)
daily['next_day_profitable'] = (daily['next_day_pnl'] > 0).astype(int)

model_df = daily.dropna(subset=['next_day_pnl', 'sentiment_value']).copy()
print(f"Rows available: {model_df.shape[0]}")
print(f"Base rate (next-day profitable): {model_df['next_day_profitable'].mean()*100:.2f}%")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, confusion_matrix, roc_curve, auc as sklearn_auc

feature_cols = ['sentiment_value', 'trade_count', 'avg_size_usd', 'win_rate', 'pnl_std_today', 'daily_pnl']
X = model_df[feature_cols]
y = model_df['next_day_profitable']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

rf = RandomForestClassifier(n_estimators=200, max_depth=5, random_state=42, class_weight='balanced')
rf.fit(X_train, y_train)

pred_rf = rf.predict(X_test)
proba_rf = rf.predict_proba(X_test)[:, 1]

print(f"Accuracy: {accuracy_score(y_test, pred_rf):.3f}")
print(f"ROC AUC: {roc_auc_score(y_test, proba_rf):.3f}")
print(classification_report(y_test, pred_rf))

importance_df = pd.DataFrame({'feature': feature_cols, 'importance': rf.feature_importances_}).sort_values('importance', ascending=False)
baseline_acc = max(y_test.mean(), 1 - y_test.mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

fig, axes = plt.subplots(1, 3, figsize=(18, 5.5))

# 1. Feature importance
sns.barplot(data=importance_df, x='importance', y='feature', hue='feature',
            palette='viridis', legend=False, ax=axes[0])
axes[0].set_title('Feature Importance (Random Forest)', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Importance')
axes[0].set_ylabel('')

# 2. ROC curve
fpr, tpr, _ = roc_curve(y_test, proba_rf)
auc_score = sklearn_auc(fpr, tpr)
axes[1].plot(fpr, tpr, color='#1976d2', linewidth=2, label=f'Random Forest (AUC={auc_score:.3f})')
axes[1].plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random guess (AUC=0.5)')
axes[1].set_title('ROC Curve: Predicting Next-Day Profitability', fontsize=12, fontweight='bold')
axes[1].set_xlabel('False Positive Rate')
axes[1].set_ylabel('True Positive Rate')
axes[1].legend()

# 3. Confusion matrix
cm = confusion_matrix(y_test, pred_rf)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[2],
            xticklabels=['Not Profitable', 'Profitable'],
            yticklabels=['Not Profitable', 'Profitable'])
axes[2].set_title('Confusion Matrix (Random Forest)', fontsize=12, fontweight='bold')
axes[2].set_xlabel('Predicted')
axes[2].set_ylabel('Actual')

model_acc = cm.trace() / cm.sum()
fig.suptitle(f'Next-Day Trader Profitability Model  |  Model Acc: {model_acc:.1%}  vs  Baseline: {baseline_acc:.1%}',
             fontsize=14, fontweight='bold', y=1.04)

plt.tight_layout()
plt.show()
```
